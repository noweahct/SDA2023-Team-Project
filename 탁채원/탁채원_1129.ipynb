### solution 2: 해당 행의 다른 열을 보고 OPS가 계산할 수 없는 상황인지 확인하고, 계산할 수 없으면 빼기
    - SLG: 타수가 0이면 계산 불가 -> 그대로 NAN 처리
    - OBP: 타수가 0이면 계산 불가 -> 그대로 NAN 처리 (OBP nan이면 타수가 모두 0임을 확인하고)
    - SLG나 OBP 중 NAN이 하나만 있으면 나머지 하나 값이 OPS가 되고, 계산 후 남은 행 중 OPS가 NAN이면 그 행은 제외한다

#### SLG = ( 1루타 개수 * 1 + 2루타 개수 * 2 + 3루타 개수 * 3 + 홈런 개수 * 4 ) / 타수
-> SLG = ( (H - (2B+3B+HR)) + 2 * 2B + 3 * 3B + 4 * HR ) / AB

```python
SLG_null = pre_season[pre_season['SLG'].isnull()]
SLG_null
```

```python
SLG_AB_zero = SLG_null[SLG_null['AB'] == 0]
SLG_AB_zero
```

```python
SLG_AB_not_zero = SLG_null[SLG_null['AB'] != 0]
SLG_AB_not_zero 
```

```python
print(SLG_null.shape[0])
print(SLG_AB_zero.shape[0])
print(SLG_AB_not_zero.shape[0])
```

```python
def cal_SLG(row):
    if np.isnan(row['SLG']) and row['AB'] != 0:
        slg = ((row['H'] - (row['2B'] + row['3B'] + row['HR'])) +
               2 * row['2B'] + 3 * row['3B'] + 4 * row['HR']) / row['AB']
        return slg
    else:
        return row['SLG']

SLG_null['SLG'] = SLG_null.apply(cal_SLG, axis=1)
SLG_null

# pre_season에는 계산할 게 없음
```
